```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
from transformers import ViTForImageClassification, ViTImageProcessor, BitsAndBytesConfig
import torchvision
from peft import LoraConfig, get_peft_model, PeftModel, prepare_model_for_kbit_training
from accelerate import Accelerator
import evaluate
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
from tqdm import tqdm
import json
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import time

import warnings
warnings.filterwarnings('ignore')

# Device configuration for MPS
def get_device():
    if torch.backends.mps.is_available():
        return torch.device("mps")
    elif torch.cuda.is_available():
        return torch.device("cuda")
    else:
        return torch.device("cpu")

device = get_device()
print(f"Using device: {device}")

# Configuration
MODEL_NAME = "google/vit-base-patch16-224-in21k"
BATCH_SIZE = 32
LEARNING_RATE = 3e-4
NUM_EPOCHS = 2
LORA_R = 16
LORA_ALPHA = 32
LORA_DROPOUT = 0.1

# QLoRA Configuration
def get_qlora_config():
    """Configure QLoRA with 4-bit quantization"""
    if device.type == "cuda":
        # Full QLoRA configuration for CUDA
        bnb_config = BitsAndBytesConfig(
            load_in_4bit=True,
            bnb_4bit_use_double_quant=True,
            bnb_4bit_quant_type="nf4",
            bnb_4bit_compute_dtype=torch.bfloat16,
        )
        return bnb_config, True
    else:
        # For MPS/CPU, we'll simulate QLoRA benefits without actual quantization
        print("QLoRA quantization not fully supported on MPS/CPU. Using optimized LoRA instead.")
        return None, False

class SimpleDataset(torch.utils.data.Dataset):
    """Simple dataset wrapper for torchvision datasets using ViT processor"""
    
    def __init__(self, torchvision_dataset, processor):
        self.dataset = torchvision_dataset
        self.processor = processor
    
    def __len__(self):
        return len(self.dataset)
    
    def __getitem__(self, idx):
        image, label = self.dataset[idx]
        
        # Convert grayscale to RGB if needed
        if image.mode != 'RGB':
            image = image.convert('RGB')
        
        # Use the ViT processor directly
        inputs = self.processor(images=image, return_tensors="pt")
        
        return {
            'pixel_values': inputs['pixel_values'].squeeze(0),
            'labels': torch.tensor(label, dtype=torch.long)
        }

def get_model_size(model_path):
    """Calculate model/adapter size in MB"""
    if os.path.isfile(model_path):
        return os.path.getsize(model_path) / (1024 * 1024)
    elif os.path.isdir(model_path):
        total_size = 0
        for dirpath, dirnames, filenames in os.walk(model_path):
            for filename in filenames:
                filepath = os.path.join(dirpath, filename)
                total_size += os.path.getsize(filepath)
        return total_size / (1024 * 1024)
    return 0

def get_model_statistics(model):
    """Get detailed model statistics"""
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    frozen_params = total_params - trainable_params
    
    # Calculate memory usage (approximate)
    param_size_mb = total_params * 4 / (1024 * 1024)  # 4 bytes per float32 parameter
    
    return {
        'total_parameters': total_params,
        'trainable_parameters': trainable_params,
        'frozen_parameters': frozen_params,
        'trainable_percentage': (trainable_params / total_params) * 100,
        'estimated_memory_mb': param_size_mb
    }

def get_quantized_model_memory(model, is_quantized=False):
    """Estimate memory usage for quantized models"""
    stats = get_model_statistics(model)
    
    if is_quantized:
        # 4-bit quantization reduces memory by ~75%
        quantized_memory = stats['estimated_memory_mb'] * 0.25
        stats['quantized_memory_mb'] = quantized_memory
        stats['memory_savings_mb'] = stats['estimated_memory_mb'] - quantized_memory
        stats['memory_savings_percentage'] = (stats['memory_savings_mb'] / stats['estimated_memory_mb']) * 100
    
    return stats

def print_model_statistics(model, model_name="Model", is_quantized=False):
    """Print formatted model statistics"""
    stats = get_quantized_model_memory(model, is_quantized)
    
    print(f"\n{model_name} Statistics:")
    print("=" * 60)
    print(f"Total Parameters:     {stats['total_parameters']:,}")
    print(f"Trainable Parameters: {stats['trainable_parameters']:,}")
    print(f"Frozen Parameters:    {stats['frozen_parameters']:,}")
    print(f"Trainable Percentage: {stats['trainable_percentage']:.2f}%")
    print(f"Estimated Memory:     {stats['estimated_memory_mb']:.2f} MB")
    
    if is_quantized and 'quantized_memory_mb' in stats:
        print(f"Quantized Memory:     {stats['quantized_memory_mb']:.2f} MB")
        print(f"Memory Savings:       {stats['memory_savings_mb']:.2f} MB ({stats['memory_savings_percentage']:.1f}%)")

def create_dataloaders(dataset_name, processor):
    """Create train and test dataloaders using torchvision"""
    
    if dataset_name == "mnist":
        train_dataset_raw = torchvision.datasets.MNIST(
            root='./data', train=True, download=True, transform=None
        )
        test_dataset_raw = torchvision.datasets.MNIST(
            root='./data', train=False, download=True, transform=None
        )
        num_classes = 10
        
    elif dataset_name == "fashion_mnist":
        train_dataset_raw = torchvision.datasets.FashionMNIST(
            root='./data', train=True, download=True, transform=None
        )
        test_dataset_raw = torchvision.datasets.FashionMNIST(
            root='./data', train=False, download=True, transform=None
        )
        num_classes = 10
        
    elif dataset_name == "cifar10":
        train_dataset_raw = torchvision.datasets.CIFAR10(
            root='./data', train=True, download=True, transform=None
        )
        test_dataset_raw = torchvision.datasets.CIFAR10(
            root='./data', train=False, download=True, transform=None
        )
        num_classes = 10
    
    else:
        raise ValueError(f"Dataset {dataset_name} not supported")
    
    train_dataset = SimpleDataset(train_dataset_raw, processor)
    test_dataset = SimpleDataset(test_dataset_raw, processor)
    
    train_dataloader = DataLoader(
        train_dataset, batch_size=BATCH_SIZE, shuffle=True, pin_memory=False
    )
    test_dataloader = DataLoader(
        test_dataset, batch_size=BATCH_SIZE, shuffle=False, pin_memory=False
    )
    
    return train_dataloader, test_dataloader, num_classes

def setup_model_with_qlora(num_classes):
    """Setup ViT model with QLoRA adapters"""
    # Get QLoRA configuration
    bnb_config, use_quantization = get_qlora_config()
    
    # Load base model with optional quantization
    if use_quantization and bnb_config is not None:
        print("Loading model with 4-bit quantization...")
        base_model = ViTForImageClassification.from_pretrained(
            MODEL_NAME,
            num_labels=num_classes,
            ignore_mismatched_sizes=True,
            quantization_config=bnb_config,
            torch_dtype=torch.bfloat16,
        )
        # Prepare model for k-bit training
        base_model = prepare_model_for_kbit_training(base_model)
        print_model_statistics(base_model, "Quantized Base ViT Model", is_quantized=True)
    else:
        print("Loading model without quantization (MPS/CPU mode)...")
        base_model = ViTForImageClassification.from_pretrained(
            MODEL_NAME,
            num_labels=num_classes,
            ignore_mismatched_sizes=True,
            torch_dtype=torch.float32
        )
        print_model_statistics(base_model, "Base ViT Model", is_quantized=False)
    
    # Configure QLoRA/LoRA
    lora_config = LoraConfig(
        r=LORA_R,
        lora_alpha=LORA_ALPHA,
        target_modules=["query", "value", "key", "dense"],  # More modules for QLoRA
        lora_dropout=LORA_DROPOUT,
        bias="none",
        modules_to_save=["classifier"],
    )
    
    # Apply LoRA to model
    model = get_peft_model(base_model, lora_config)
    
    model_type = "QLoRA" if use_quantization else "Optimized LoRA"
    print_model_statistics(model, f"ViT with {model_type}", is_quantized=use_quantization)
    
    return model, use_quantization

def evaluate_model_comprehensive(model, test_dataloader, class_names, dataset_name):
    """Comprehensive model evaluation with detailed metrics"""
    model.eval()
    all_predictions = []
    all_labels = []
    all_logits = []
    inference_times = []
    
    print(f"\n Evaluating {dataset_name.upper()} model...")
    
    with torch.no_grad():
        for batch in tqdm(test_dataloader, desc="Evaluating"):
            start_time = time.time()
            
            pixel_values = batch['pixel_values'].to(device)
            labels = batch['labels'].to(device)
            
            outputs = model(pixel_values=pixel_values)
            predictions = torch.argmax(outputs.logits, dim=-1)
            
            inference_times.append(time.time() - start_time)
            
            all_predictions.extend(predictions.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
            all_logits.extend(outputs.logits.cpu().numpy())
    
    # Calculate metrics
    accuracy = np.mean(np.array(all_predictions) == np.array(all_labels))
    avg_inference_time = np.mean(inference_times)
    
    # Classification report
    report = classification_report(
        all_labels, all_predictions, 
        target_names=class_names, 
        output_dict=True
    )
    
    # Confusion matrix
    cm = confusion_matrix(all_labels, all_predictions)
    
    metrics = {
        'accuracy': accuracy,
        'precision_macro': report['macro avg']['precision'],
        'recall_macro': report['macro avg']['recall'],
        'f1_macro': report['macro avg']['f1-score'],
        'precision_weighted': report['weighted avg']['precision'],
        'recall_weighted': report['weighted avg']['recall'],
        'f1_weighted': report['weighted avg']['f1-score'],
        'avg_inference_time': avg_inference_time,
        'confusion_matrix': cm.tolist(),
        'classification_report': report
    }
    
    return metrics, all_predictions, all_labels

def plot_confusion_matrix(cm, class_names, dataset_name):
    """Plot confusion matrix"""
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f'{dataset_name.upper()} - QLoRA Confusion Matrix')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.tight_layout()
    plt.show()

def print_classification_metrics(metrics, dataset_name):
    """Print detailed classification metrics"""
    print(f"\n {dataset_name.upper()} QLoRA Classification Metrics:")
    print("=" * 50)
    print(f"Accuracy:           {metrics['accuracy']:.4f}")
    print(f"Precision (Macro):  {metrics['precision_macro']:.4f}")
    print(f"Recall (Macro):     {metrics['recall_macro']:.4f}")
    print(f"F1-Score (Macro):   {metrics['f1_macro']:.4f}")
    print(f"Precision (Weighted): {metrics['precision_weighted']:.4f}")
    print(f"Recall (Weighted):    {metrics['recall_weighted']:.4f}")
    print(f"F1-Score (Weighted):  {metrics['f1_weighted']:.4f}")
    print(f"Avg Inference Time: {metrics['avg_inference_time']:.4f} sec/batch")
    
    # Per-class metrics
    report = metrics['classification_report']
    print(f"\n Per-Class Metrics:")
    print("-" * 70)
    print(f"{'Class':<15} {'Precision':<12} {'Recall':<12} {'F1-Score':<12} {'Support':<10}")
    print("-" * 70)
    
    for class_name in report.keys():
        if class_name not in ['accuracy', 'macro avg', 'weighted avg']:
            class_metrics = report[class_name]
            print(f"{class_name:<15} {class_metrics['precision']:<12.4f} "
                  f"{class_metrics['recall']:<12.4f} {class_metrics['f1-score']:<12.4f} "
                  f"{class_metrics['support']:<10.0f}")

def train_model(model, train_dataloader, test_dataloader, dataset_name, class_names, is_quantized):
    """Train the model with QLoRA adapters"""
    accelerator = Accelerator(device_placement=True)
    
    # Use different optimizer settings for quantized models
    if is_quantized:
        optimizer = torch.optim.AdamW(model.parameters(), lr=LEARNING_RATE, weight_decay=0.01)
    else:
        optimizer = torch.optim.AdamW(model.parameters(), lr=LEARNING_RATE)
    
    model, optimizer, train_dataloader, test_dataloader = accelerator.prepare(
        model, optimizer, train_dataloader, test_dataloader
    )
    
    model.train()
    best_accuracy = 0.0
    training_history = {'epochs': [], 'train_loss': [], 'val_accuracy': []}
    
    for epoch in range(NUM_EPOCHS):
        total_loss = 0
        progress_bar = tqdm(train_dataloader, desc=f"Epoch {epoch+1}/{NUM_EPOCHS}")
        
        for batch in progress_bar:
            optimizer.zero_grad()
            
            pixel_values = batch['pixel_values']
            labels = batch['labels']
            
            outputs = model(pixel_values=pixel_values, labels=labels)
            loss = outputs.loss
            
            accelerator.backward(loss)
            
            # Gradient clipping for stability with quantized models
            if is_quantized:
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            
            optimizer.step()
            
            total_loss += loss.item()
            progress_bar.set_postfix({'loss': loss.item()})
        
        avg_loss = total_loss / len(train_dataloader)
        
        # Quick validation
        model.eval()
        val_correct = 0
        val_total = 0
        
        with torch.no_grad():
            for batch in test_dataloader:
                pixel_values = batch['pixel_values']
                labels = batch['labels']
                
                outputs = model(pixel_values=pixel_values)
                predictions = torch.argmax(outputs.logits, dim=-1)
                
                predictions = accelerator.gather(predictions)
                labels = accelerator.gather(labels)
                
                val_correct += (predictions == labels).float().sum().item()
                val_total += labels.size(0)
        
        val_accuracy = val_correct / val_total
        
        print(f"Epoch {epoch+1} - Loss: {avg_loss:.4f}, Val Accuracy: {val_accuracy:.4f}")
        
        # Save training history
        training_history['epochs'].append(epoch + 1)
        training_history['train_loss'].append(avg_loss)
        training_history['val_accuracy'].append(val_accuracy)
        
        # Save best model
        if val_accuracy > best_accuracy:
            best_accuracy = val_accuracy
            unwrapped_model = accelerator.unwrap_model(model)
            
            # Comprehensive evaluation on best model
            metrics, predictions, true_labels = evaluate_model_comprehensive(
                unwrapped_model, test_dataloader, class_names, dataset_name
            )
            
            save_qlora_adapter_and_config(unwrapped_model, dataset_name, metrics, training_history, is_quantized)
        
        model.train()
    
    print(f"Best accuracy for {dataset_name}: {best_accuracy:.4f}")
    return model, best_accuracy, metrics

def save_qlora_adapter_and_config(model, dataset_name, metrics, training_history, is_quantized):
    """Save QLoRA adapters, configuration, and comprehensive metrics"""
    adapter_dir = f"qlora_adapters_{dataset_name}"
    
    # Save QLoRA adapters
    model.save_pretrained(adapter_dir)
    
    # Get adapter size
    adapter_size = get_model_size(adapter_dir)
    model_stats = get_quantized_model_memory(model, is_quantized)
    
    # Save comprehensive configuration
    config = {
        "dataset_name": dataset_name,
        "model_name": MODEL_NAME,
        "num_classes": 10,
        "method": "QLoRA" if is_quantized else "Optimized LoRA",
        "quantized": is_quantized,
        "qlora_config": {
            "r": LORA_R,
            "alpha": LORA_ALPHA,
            "dropout": LORA_DROPOUT,
            "target_modules": ["query", "value", "key", "dense"],
            "quantization": "4-bit" if is_quantized else "none"
        },
        "training_config": {
            "batch_size": BATCH_SIZE,
            "learning_rate": LEARNING_RATE,
            "num_epochs": NUM_EPOCHS
        },
        "model_statistics": model_stats,
        "adapter_size_mb": adapter_size,
        "metrics": metrics,
        "training_history": training_history
    }
    
    with open(f"{adapter_dir}/qlora_comprehensive_config.json", "w") as f:
        json.dump(config, f, indent=2)
    
    method_name = "QLoRA" if is_quantized else "Optimized LoRA"
    print(f"{method_name} adapters and comprehensive config saved to {adapter_dir}")
    print(f"Adapter size: {adapter_size:.2f} MB")
    
    if is_quantized and 'memory_savings_mb' in model_stats:
        print(f"Memory savings: {model_stats['memory_savings_mb']:.2f} MB ({model_stats['memory_savings_percentage']:.1f}%)")

def load_model_with_qlora_adapters(dataset_name):
    """Load base model and apply saved QLoRA adapters"""
    adapter_dir = f"qlora_adapters_{dataset_name}"
    
    if not os.path.exists(adapter_dir):
        raise FileNotFoundError(f"QLoRA adapter directory {adapter_dir} not found.")
    
    # Load configuration
    with open(f"{adapter_dir}/qlora_comprehensive_config.json", "r") as f:
        config = json.load(f)
    
    is_quantized = config.get('quantized', False)
    method_name = config.get('method', 'QLoRA')
    
    print(f"Loading {dataset_name} {method_name} model (Accuracy: {config['metrics']['accuracy']:.4f})")
    
    # Load base model with appropriate configuration
    if is_quantized and device.type == "cuda":
        bnb_config, _ = get_qlora_config()
        base_model = ViTForImageClassification.from_pretrained(
            config["model_name"],
            num_labels=config["num_classes"],
            ignore_mismatched_sizes=True,
            quantization_config=bnb_config,
            torch_dtype=torch.bfloat16,
        )
    else:
        base_model = ViTForImageClassification.from_pretrained(
            config["model_name"],
            num_labels=config["num_classes"],
            ignore_mismatched_sizes=True,
            torch_dtype=torch.float32
        )
    
    # Load QLoRA adapters
    model = PeftModel.from_pretrained(base_model, adapter_dir)
    model = model.to(device)
    model.eval()
    
    return model, config

def display_comprehensive_qlora_statistics():
    """Display comprehensive statistics for all trained QLoRA adapters"""
    print(f"\n{'='*80}")
    print("COMPREHENSIVE QLORA MODEL AND ADAPTER STATISTICS")
    print(f"{'='*80}")
    
    adapters = []
    for item in os.listdir('.'):
        if item.startswith('qlora_adapters_') and os.path.isdir(item):
            dataset_name = item.replace('qlora_adapters_', '')
            config_path = os.path.join(item, 'qlora_comprehensive_config.json')
            
            if os.path.exists(config_path):
                with open(config_path, 'r') as f:
                    config = json.load(f)
                adapters.append((dataset_name, config))
    
    if not adapters:
        print("No trained QLoRA adapters found.")
        return
    
    # Model Statistics Table
    print(f"\n QLORA MODEL ARCHITECTURE STATISTICS")
    print("-" * 100)
    print(f"{'Dataset':<15} {'Method':<15} {'Total Params':<15} {'Trainable':<15} {'Trainable %':<12} {'Size (MB)':<10}")
    print("-" * 100)
    
    for dataset_name, config in adapters:
        stats = config['model_statistics']
        method = config.get('method', 'QLoRA')
        print(f"{dataset_name.upper():<15} {method:<15} {stats['total_parameters']:<15,} "
              f"{stats['trainable_parameters']:<15,} {stats['trainable_percentage']:<12.2f} "
              f"{config['adapter_size_mb']:<10.2f}")
    
    # Memory Efficiency Table
    print(f"\n MEMORY EFFICIENCY (QUANTIZATION BENEFITS)")
    print("-" * 80)
    print(f"{'Dataset':<15} {'Original (MB)':<15} {'Quantized (MB)':<15} {'Savings (MB)':<15} {'Savings %':<12}")
    print("-" * 80)
    
    for dataset_name, config in adapters:
        stats = config['model_statistics']
        original_mem = stats.get('estimated_memory_mb', 0)
        quantized_mem = stats.get('quantized_memory_mb', original_mem)
        savings_mb = stats.get('memory_savings_mb', 0)
        savings_pct = stats.get('memory_savings_percentage', 0)
        
        print(f"{dataset_name.upper():<15} {original_mem:<15.2f} {quantized_mem:<15.2f} "
              f"{savings_mb:<15.2f} {savings_pct:<12.1f}")
    
    # Performance Metrics Table
    print(f"\n QLORA PERFORMANCE METRICS")
    print("-" * 100)
    print(f"{'Dataset':<15} {'Accuracy':<10} {'F1-Macro':<10} {'F1-Weighted':<12} {'Precision':<11} {'Recall':<10} {'Inf.Time':<10}")
    print("-" * 100)
    
    for dataset_name, config in adapters:
        metrics = config['metrics']
        print(f"{dataset_name.upper():<15} {metrics['accuracy']:<10.4f} "
              f"{metrics['f1_macro']:<10.4f} {metrics['f1_weighted']:<12.4f} "
              f"{metrics['precision_macro']:<11.4f} {metrics['recall_macro']:<10.4f} "
              f"{metrics['avg_inference_time']:<10.4f}")
    
    # QLoRA Efficiency Comparison
    print(f"\n QLORA EFFICIENCY COMPARISON")
    print("-" * 80)
    print(f"{'Dataset':<15} {'Adapter Size':<15} {'Compression':<15} {'Params Saved':<15} {'Quantized':<12}")
    print("-" * 80)
    
    base_model_size = 330  # Approximate size of ViT-Base in MB
    
    for dataset_name, config in adapters:
        adapter_size = config['adapter_size_mb']
        compression_ratio = base_model_size / adapter_size if adapter_size > 0 else 0
        params_saved = (1 - config['model_statistics']['trainable_percentage'] / 100) * 100
        is_quantized = "Yes" if config.get('quantized', False) else "No"
        
        print(f"{dataset_name.upper():<15} {adapter_size:<15.2f} "
              f"{compression_ratio:<15.1f}x {params_saved:<15.1f}% {is_quantized:<12}")

def plot_training_curves():
    """Plot training curves for all QLoRA datasets"""
    adapters = []
    for item in os.listdir('.'):
        if item.startswith('qlora_adapters_') and os.path.isdir(item):
            dataset_name = item.replace('qlora_adapters_', '')
            config_path = os.path.join(item, 'qlora_comprehensive_config.json')
            
            if os.path.exists(config_path):
                with open(config_path, 'r') as f:
                    config = json.load(f)
                adapters.append((dataset_name, config))
    
    if not adapters:
        print("No QLoRA training history found.")
        return
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    
    for dataset_name, config in adapters:
        history = config['training_history']
        method = config.get('method', 'QLoRA')
        label = f"{dataset_name.upper()} ({method})"
        
        ax1.plot(history['epochs'], history['train_loss'], 
                marker='o', label=label)
        ax2.plot(history['epochs'], history['val_accuracy'], 
                marker='s', label=label)
    
    ax1.set_title('QLoRA Training Loss')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.legend()
    ax1.grid(True)
    
    ax2.set_title('QLoRA Validation Accuracy')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy')
    ax2.legend()
    ax2.grid(True)
    
    plt.tight_layout()
    plt.show()

def plot_inference_results(pixel_values, labels, predictions, dataset_name, class_names, num_samples=10):
    """Plot QLoRA inference results"""
    fig, axes = plt.subplots(2, 5, figsize=(15, 6))
    fig.suptitle(f'{dataset_name.upper()} - QLoRA Inference Results', fontsize=16)
    
    num_samples = min(num_samples, len(pixel_values))
    
    for i in range(num_samples):
        row = i // 5
        col = i % 5
        
        # Convert tensor to numpy and reshape for display
        image = pixel_values[i].numpy().transpose(1, 2, 0)
        
        # Denormalize image
        mean = np.array([0.485, 0.456, 0.406])
        std = np.array([0.229, 0.224, 0.225])
        image = image * std + mean
        image = np.clip(image, 0, 1)
        
        # For grayscale datasets, convert to grayscale for display
        if dataset_name in ['mnist', 'fashion_mnist']:
            image = np.mean(image, axis=2)
            axes[row, col].imshow(image, cmap='gray')
        else:
            axes[row, col].imshow(image)
        
        true_label = labels[i].item()
        pred_label = predictions[i].item()
        
        color = 'green' if true_label == pred_label else 'red'
        axes[row, col].set_title(f'True: {class_names[true_label]}\nPred: {class_names[pred_label]}', 
                                color=color, fontsize=10)
        axes[row, col].axis('off')
    
    plt.tight_layout()
    plt.show()

def run_inference_on_dataset(dataset_name, class_names, processor):
    """Run comprehensive inference on a specific dataset using QLoRA"""
    print(f"\n Running comprehensive QLoRA inference on {dataset_name.upper()}")
    
    try:
        model, config = load_model_with_qlora_adapters(dataset_name)
        _, test_dataloader, _ = create_dataloaders(dataset_name, processor)
        
        # Load saved metrics or run evaluation
        if 'metrics' in config:
            metrics = config['metrics']
            print(f"Using saved metrics from training...")
        else:
            print(f"Running fresh evaluation...")
            metrics, _, _ = evaluate_model_comprehensive(
               model, test_dataloader, class_names, dataset_name
            )
        
        # Print detailed metrics
        print_classification_metrics(metrics, dataset_name)
        
        # Plot confusion matrix
        cm = np.array(metrics['confusion_matrix'])
        plot_confusion_matrix(cm, class_names, dataset_name)
        
        # Plot inference examples
        test_batch = next(iter(test_dataloader))
        with torch.no_grad():
           pixel_values = test_batch['pixel_values'].to(device)
           outputs = model(pixel_values=pixel_values)
           predictions = torch.argmax(outputs.logits, dim=-1)
        
        plot_inference_results(
           test_batch['pixel_values'].cpu(),
           test_batch['labels'].cpu(),
           predictions.cpu(),
           dataset_name,
           class_names
        )
        
        return model, metrics
       
    except Exception as e:
        print(f" Error with {dataset_name}: {e}")
        return None, None

def compare_lora_vs_qlora():
    """Compare LoRA vs QLoRA results if both exist"""
    print(f"\n{'='*80}")
    print(" LORA VS QLORA COMPARISON")
    print(f"{'='*80}")
    
    # Check for both LoRA and QLoRA adapters
    lora_adapters = {}
    qlora_adapters = {}
    
    for item in os.listdir('.'):
       if item.startswith('lora_adapters_') and os.path.isdir(item):
           dataset_name = item.replace('lora_adapters_', '')
           config_path = os.path.join(item, 'comprehensive_config.json')
           if os.path.exists(config_path):
               with open(config_path, 'r') as f:
                   lora_adapters[dataset_name] = json.load(f)
       
       elif item.startswith('qlora_adapters_') and os.path.isdir(item):
           dataset_name = item.replace('qlora_adapters_', '')
           config_path = os.path.join(item, 'qlora_comprehensive_config.json')
           if os.path.exists(config_path):
               with open(config_path, 'r') as f:
                   qlora_adapters[dataset_name] = json.load(f)
    
    # Find common datasets
    common_datasets = set(lora_adapters.keys()) & set(qlora_adapters.keys())
   
    if not common_datasets:
        print("No common datasets found between LoRA and QLoRA experiments.")
        return
   
    print(f"\n PERFORMANCE COMPARISON")
    print("-" * 80)
    print(f"{'Dataset':<15} {'LoRA Acc':<12} {'QLoRA Acc':<12} {'Diff':<8} {'LoRA Time':<12} {'QLoRA Time':<12}")
    print("-" * 80)
   
    for dataset in sorted(common_datasets):
        lora_acc = lora_adapters[dataset]['metrics']['accuracy']
        qlora_acc = qlora_adapters[dataset]['metrics']['accuracy']
        acc_diff = qlora_acc - lora_acc
        lora_time = lora_adapters[dataset]['metrics']['avg_inference_time']
        qlora_time = qlora_adapters[dataset]['metrics']['avg_inference_time']
        
        print(f"{dataset.upper():<15} {lora_acc:<12.4f} {qlora_acc:<12.4f} "
             f"{acc_diff:<8.4f} {lora_time:<12.4f} {qlora_time:<12.4f}")
    
    print(f"\n MEMORY EFFICIENCY COMPARISON")
    print("-" * 80)
    print(f"{'Dataset':<15} {'LoRA Size':<12} {'QLoRA Size':<12} {'Memory Savings':<15} {'Compression':<12}")
    print("-" * 80)
   
    for dataset in sorted(common_datasets):
        lora_size = lora_adapters[dataset]['adapter_size_mb']
        qlora_size = qlora_adapters[dataset]['adapter_size_mb']
        
        qlora_stats = qlora_adapters[dataset]['model_statistics']
        memory_savings = qlora_stats.get('memory_savings_mb', 0)
        compression_ratio = lora_size / qlora_size if qlora_size > 0 else 1
        
        print(f"{dataset.upper():<15} {lora_size:<12.2f} {qlora_size:<12.2f} "
             f"{memory_savings:<15.2f} {compression_ratio:<12.2f}x")

def main():
    # Check if quantization is available
    bnb_config, use_quantization = get_qlora_config()
    if use_quantization:
       print("4-bit quantization enabled (CUDA detected)")
    else:
       print("Running optimized LoRA (MPS/CPU - quantization not available)")
    
    # Initialize processor
    processor = ViTImageProcessor.from_pretrained(MODEL_NAME)
    
    # Dataset configurations
    # Dataset configurations
    datasets_info = [
        ("mnist", [str(i) for i in range(10)]),
        (
            "fashion_mnist",
            [
                "T-shirt/top",
                "Trouser",
                "Pullover",
                "Dress",
                "Coat",
                "Sandal",
                "Shirt",
                "Sneaker",
                "Bag",
                "Ankle boot",
            ],
        ),
        (
            "cifar10",
            [
                "airplane",
                "automobile",
                "bird",
                "cat",
                "deer",
                "dog",
                "frog",
                "horse",
                "ship",
                "truck",
            ],
        ),
    ]
    
    print(f"\n{'='*80}")
    print(" PHASE 1: TRAINING QLORA ADAPTERS FOR ALL DATASETS")
    print(f"{'='*80}")

   # Train all datasets
    for dataset_name, class_names in datasets_info:
        print(f"\n{'='*50}")
        print(f"Training QLoRA on {dataset_name.upper()} dataset")
        print(f"{'='*50}")
        
        train_dataloader, test_dataloader, num_classes = create_dataloaders(dataset_name, processor)
        model, is_quantized = setup_model_with_qlora(num_classes)
        
        _, _, metrics = train_model(model, train_dataloader, test_dataloader, 
                                 dataset_name, class_names, is_quantized)
        
        # Print detailed metrics
        print_classification_metrics(metrics, dataset_name)
        
        # Plot confusion matrix
        cm = np.array(metrics['confusion_matrix'])
        plot_confusion_matrix(cm, class_names, dataset_name)
        
        # Clear memory
        del model
        torch.cuda.empty_cache() if torch.cuda.is_available() else None
        if device.type == "mps":
           torch.mps.empty_cache()
   
    print(f"\n{'='*80}")
    print("PHASE 2: COMPREHENSIVE QLORA INFERENCE AND ANALYSIS")
    print(f"{'='*80}")
   
    # Run inference on all trained models
    for dataset_name, class_names in datasets_info:
        run_inference_on_dataset(dataset_name, class_names, processor)
   
    print(f"\n{'='*80}")
    print(" PHASE 3: FINAL COMPREHENSIVE QLORA STATISTICS")
    print(f"{'='*80}")
   
   # Display comprehensive statistics
    display_comprehensive_qlora_statistics()
    
    # Plot training curves
    plot_training_curves()
    
    # Compare with regular LoRA if available
    compare_lora_vs_qlora()

if __name__ == "__main__":
   main()
```
